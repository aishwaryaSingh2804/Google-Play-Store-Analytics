## Google Play Store Apps Analysis

### Business Problem

The Google Play Store contains millions of applications across different categories. Understanding what contributes to successful applications can help developers and businesses make better decisions.

### Objectives

- Analyze the distribution of apps across categories.
- Identify factors affecting ratings.
- Study installs and pricing.
- Examine relationships between app size, category and popularity.
- Discover trends in user reviews.
- Provide business recommendations.
