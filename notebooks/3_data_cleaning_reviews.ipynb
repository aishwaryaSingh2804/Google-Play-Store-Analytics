### CLEANING REVIEWS DATASET
#### Process
    1. Analyzing and removing null values
    2. Analyzing duplicate values and deciding whether to drop or keep them
    3. Checking the quality of data in each column

```python
import pandas as pd
```

```python
reviews = pd.read_csv('../datasets/reviews.csv')
```

```python
# if there is no review, then there is no point of sentiment analysis or any other column
reviews.loc[lambda x : x['Translated_Review'].isna()].head(10)
reviews.loc[lambda x : x['Translated_Review'].isna()].tail(10)
```

```python
# REMOVING NULL REVIEWS
reviews = reviews.dropna(subset=['Translated_Review'])
print('Null values\n',reviews.isna().sum())
# so null values successfully removed
print('\nNon-null rows :\t',reviews.shape[0])
```

```python
# REMOVING DUPLICATES

reviews[reviews.sort_values(by=['App','Translated_Review']).duplicated(subset=['App', 'Translated_Review'], keep=False)].tail(6)

# since mulitple users can give the same review, i wont drop duplicates
```

```python
#CHECKING QUALITY OF VALUES IN EACH COLUMN
print('Values in Sentiment Column\n',reviews['Sentiment'].unique(),'\n')
print('Checking Quality of Values in Sentiment Polarity and Sentiment Subjectivity Column')
reviews.agg({'Sentiment_Polarity':['min','max','mean'], 'Sentiment_Subjectivity':['min','max','mean']})
```

```python
reviews.to_csv('cleaned_reviews.csv', index=False)
```
