### CLEANING APPS DATASET
#### Process
1. Removing duplicates
2. Modifying data types of columns
3. Handling null values

```python
import pandas as pd
import numpy as np
apps = pd.read_csv('../datasets/apps.csv')
reviews = pd.read_csv('../datasets/reviews.csv')
```

```python
# ANALYZING DUPLICATE ROWS
# duplicate rows
apps[apps.duplicated()].sort_values(['App', 'Last Updated'])
```

```python
# REMOVING DUPLICATES

# dropping the duplicate rows and only keeping the last row of the dataset sorted by reviews
apps = apps.sort_values('Reviews').drop_duplicates(subset=['App'], keep='last')

apps.duplicated().sum()
```

```python
# CONVERTING DATA TYPES TO DESIRED DATA TYPES

print('Numeric Values Counts in Reviews')
print(apps['Reviews'].str.isnumeric().value_counts())
# so one row has non numeric review

print(apps[apps['Reviews'].str.isnumeric() == False])
# on inspection this row is malformed so we need to drop this

apps = apps[apps['Reviews'].str.isnumeric()]

print('\n\nNumeric Values Counts in Reviews')
print(apps['Reviews'].str.isnumeric().value_counts())

#now all rows in Reviews have numeric values and can be converted to int
apps['Reviews'] = apps['Reviews'].astype(int)
```

During data validation, one record was found with a non-numeric value in the Reviews column. Inspection showed that the entire row was malformed due to a column misalignment rather than a simple formatting issue. Since the record could not be reliably corrected, it was removed from the dataset.

```python
#CONVERTING ALL VALUES IN INSTALLS COLUMN TO NUMERIC VALUES

clean_installs = (
    apps['Installs']
        .str.replace(',', '', regex=False)
        .str.replace('+', '', regex=False)
)

print('Rows with numeric values in Installs\n',clean_installs.str.isnumeric().value_counts())

# since all rows are numeric, data type of Installs can be changed

# Clean Installs
apps['Installs'] = (
    apps['Installs']
        .str.replace(',', '', regex=False)
        .str.replace('+', '', regex=False)
        .astype(int)
)
```

```python
# CONVERTING PRICE TO FLOAT
apps['Price'].unique()

apps['Price'].str.replace('$','').unique()

apps['Price'] = apps['Price'].str.replace('$','')

apps['Price'] = apps['Price'].astype(float)
```

```python
apps['Last Updated'] = pd.to_datetime(
    apps['Last Updated'],
    format='%B %d, %Y'
)
```

```python
# exploring statistics of reviews on basis of null ratings

print('No Ratings\n',apps.loc[apps['Rating'].isna(), 'Reviews'].describe())
print('\nWith Ratings\n',apps.loc[apps['Rating'].notna(), 'Reviews'].describe())

apps.loc[apps['Rating'].isna(), 'Reviews'].value_counts()
```

```python
apps.dropna(subset=['Rating'], inplace=True)
```

### Handling Missing Ratings

The missing values in the `Rating` column were investigated before applying any imputation technique.

Analysis showed that applications without ratings had a median of only 1 review, whereas applications with ratings had a median of 3,004 reviews. Additionally, over 40% of the unrated apps had zero reviews.

This indicates that the missing ratings are not random; rather, these applications had not received enough user feedback to generate a rating. Therefore, imputing missing ratings with the mean or median would introduce artificial values and bias the analysis.

To preserve data quality, rows with missing ratings were removed.

```python
print(apps.isna().sum())
```

```python
apps.loc[apps['Current Ver'].isna() | apps['Android Ver'].isna(),  ['App', 'Current Ver', 'Android Ver', 'Category', 'Rating', 'Reviews', 'Installs']]
```

The Current Ver and Android Ver columns contained only 4 and 2 missing values respectively. Since these columns were not central to the subsequent analyses and reliable values could not be inferred, the missing values were retained to avoid introducing inaccurate information.
